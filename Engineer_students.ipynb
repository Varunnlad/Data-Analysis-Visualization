```python
import pandas as pd

df = pd.read_csv("students.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

After Analysing we understand that some categorical Terms should convert to category datatypes


```python
df['Gender']= df['Gender'].astype('category')
df['Placement Status']= df['Placement Status'].astype('category')
df['Branch']= df['Branch'].astype('category')
df['Internship Done']= df['Internship Done'].astype('category')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Placement Status'] = df['Placement Status'].str.strip().str.lower()
df['Internship Done'] = df['Internship Done'].str.strip().str.lower()
```

# Lets Visualize some relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="Placement Status")
plt.title("Placement Distributiuon")
```

```python
sns.boxplot(data=df, x="Placement Status", y="Average GPA")
plt.title("Placement vs CGPA")
```

```python
sns.boxplot(data=df, x='Placement Status', y='Backlogs')
plt.title('Backlogs by Placement')
```

```python
sns.boxplot(data=df, x='Placement Status', y='Attendance (%)')
plt.title('Attendance by Placement')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df , x="Branch", hue = "Placement Status")
plt.xticks(rotation=45)
plt.title("Placement Distribution by Branch")
```

```python
sem_cols = [col for col in df.columns if 'Sem' in col]
df[sem_cols].T.plot( legend=False,title="Semester GPA Trends (All Students Sampled)")
plt.xticks(rotation=45)
```

```python
df['GPA Growth'] = df['Sem8 GPA'] - df['Sem1 GPA']
sns.histplot(df['GPA Growth'])
plt.title('GPA Growth Distribution')
```

```python
sns.countplot(data=df, x='Internship Done', hue='Placement Status')
plt.title("Internship vs Placement")
```

```python
pivot = pd.crosstab(df['Internship Domain'], df['Placement Domain'])
sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
plt.title("Internship Domain vs Placement Domain")
```

```python
import numpy as np
numeric_cols = df.select_dtypes(include=np.number)
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
```
